```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import math

def gainCR(omega, RC):
    gain = omega*RC/np.sqrt(1.0 + (omega*RC)**2)
    return gain

def gainRC(omega, RC):
    gain = 1.0/np.sqrt(1.0 + (omega*RC)**2)
    return gain

x1, x2 = 1.0, 1.0E+6
x1log, x2log = math.log10(x1), math.log10(x2)
n = 100
xlog = np.arange(x1log, x2log, (x2log - x1log)/n)
x = np.power(10.0, xlog)
tau = 5.0E+3
RC = 1.0/tau
yCR = gainCR(x, RC=RC)
yRC = gainRC(x, RC=RC)
#
plt.plot(x, yCR)
plt.plot(x, yRC, color='red')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('omega')
plt.ylabel('Gain')
plt.show()
```
