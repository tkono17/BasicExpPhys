```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0.0, 100.0, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
plt.hist(y, bins=100, range=(-1.0, 1.0))
plt.xlabel('Y')
plt.ylabel('Counts')
plt.show()
fout = open('../data/2022ALH1_sindata.dat', 'w')
for i in range(len(x)):
  fout.write('%10.5f %10.5f\n' % (x[i], y[i]) )
fout.close()
```
